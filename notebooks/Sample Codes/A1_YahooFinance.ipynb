```python
import yfinance as yf
import mplfinance as mpf
import warnings # I turnoff warning messages

# Suppress specific warning categories
warnings.simplefilter(action='ignore', category=FutureWarning) # somehow 1 msg still apears on my PC
warnings.simplefilter(action='ignore', category=UserWarning)

# Define the ticker symbol for oil (WTI/Brent)
ticker_symbol = "BZ=F"  # Brent crude oil, change if necessary

# Fetch historical data
oil_data = yf.download(ticker_symbol, start="2020-01-01", end="2024-01-01")

# Save to oil.csv
oil_data.to_csv('oil.csv')

# Plot OHLC candlestick chart
mpf.plot(oil_data, type='candle', style='charles',
         title='Oil Price OHLC Candlestick Chart',
         ylabel='Price ($)',
         figratio=(16, 9),
         volume=True)
```
